#### FEATURE TRANSFORMATION

```python
from sklearn.preprocessing import LabelEncoder
```

```python
category_mappings = {}
label_encoder = LabelEncoder()
df2 = pd.DataFrame()
for i in df.columns:
    if df[i].dtype == 'object':
        df2[i] = label_encoder.fit_transform(df[i])
        category_mappings[f'category_mapping_df2[{i}]'] = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    else:
        df2[i] = df[i]
```

```python
df['time_trends'] = 0
df['city_trends'] = 0
for i in range(df.shape[0]):
    if df['CTR_last_7_days'].iloc[i] != 0 and df['CTR_last_30_days'].iloc[i]!= 0:
        df['time_trends'].iloc[i] = df['CTR_last_7_days'].iloc[i]/df['CTR_last_30_days'].iloc[i]
    else:
        df['time_trends'].iloc[i] = 0
    if df['CTR_product_30_days'].iloc[i]!= 0 and df['CTR_plt_30_days'].iloc[i]!= 0:
        df['city_trends'].iloc[i] = df['CTR_product_30_days'].iloc[i]/df['CTR_plt_30_days'].iloc[i]
    else:
        df['city_trends'].iloc[i] = 0
```

#### FEATURE SELECTION

```python
from sklearn.model_selection import train_test_split
x = df2.drop(columns = ['is_clicked'])
y = df['is_clicked']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier()
rf.fit(x_train, y_train)

importances = rf.feature_importances_
feature_names = x.columns
feature_importances = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)
top_16_features = feature_importances['Feature'][:16].tolist()
x_train_selected = x_train[top_16_features]
x_test_selected = x_test[top_16_features]
```

```python
rf_feature_selected = RandomForestClassifier(n_estimators=250, random_state=42)
rf_feature_selected.fit(x_train_selected, y_train)
y_pred_rf = rf_feature_selected.predict(x_test_selected)
accuracy = accuracy_score(y_test, y_pred_rf)
```

#### MODEL BUILDING

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
model = RandomForestClassifier(n_estimators=300, max_depth=8, random_state=42)
model.fit(x_train_selected, y_train)
```

```python
y_pred = model.predict(x_test_selected)
y_pred_proba = model.predict_proba(x_test_selected)[:, 1]
accuracy = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)
print(f'Accuracy: {accuracy}')
print(f'ROC AUC: {roc_auc}')
print(f'Predicted probabilities: {y_pred_proba[:8]}')
```
